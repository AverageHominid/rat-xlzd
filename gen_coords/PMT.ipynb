```python
from math import sqrt, sin, cos, pi
import random
import numpy as np
```

```python
path_to_bacc = "/home/kian/bacc-xlzd-master" #change this to your path

data = np.loadtxt(path_to_bacc+"/G3/detectorConfig/pmt_array_40t.txt") #can change this to 60t depending on desired size
```

```python
r = np.sqrt(data[:,1]**2+data[:,0]**2) # radius (for PTFE holes)
a = np.arctan2(data[:,1], data[:,0]) # angle in radians
```

```python
#for the pmtinfo file
z_offset_pls = 20.0+699.8
z_offset_mns = 20.0+699.8
data_length = len(data[:,0])
write_text = ['//Generated by PMT.ipynb', '{', ' name: "PMTINFO_inner",', 'valid_begin: [0,0],', 'valid_end: [0,0],']
write_text.append('x:'+str(data[:,0].tolist()+data[:,0].tolist())+',')
write_text.append('y:'+str(data[:,1].tolist()+data[:,1].tolist())+',')
write_text.append('z:'+str([z_offset_pls for point in range(data_length)]+[-z_offset_mns for point in range(data_length)]))
write_text.append('dir_x:'+str([0.0 for point in range(2*data_length)]))
write_text.append('dir_y:'+str([0.0 for point in range(2*data_length)]))
write_text.append('dir_z:'+str([-1.0 for point in range(data_length)]+[1.0 for point in range(data_length)]))
write_text.append('type:'+str([1.0 for point in range(2*data_length)]))





write_text.append('}')
write_text = [line + '\n' for line in write_text]

file = open("PMTINFO_inner.ratdb", "a")
file.writelines(write_text)
file.close()
```
